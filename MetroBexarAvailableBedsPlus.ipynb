```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
metro_pd = pd.read_csv('../CoVID-19_Dashboard_Data.csv')
```

```python
c = b['AvailStaffedBeds'].dropna()
```

```python
d = b['TotalStaffedBeds'].dropna()
```

```python
e = b["PosPatients"].dropna()
```

```python
type(c.index[0])
```

```python
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%b %Y')
fig, ax = plt.subplots()
fig.set_size_inches(8,4)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
#ax.bar(c.index, c, width=0.8, color='g', align='center')
plt.xlabel('Date',color='black')
plt.ylabel('Number Beds/Patients',color='black')
plt.title('Metro Bexar Total/Available Staffed Hospital Beds and Patients')
plt.plot(d)
plt.plot(e, linewidth=5, markersize=12)
ax.bar(c.index,c)
ax.legend(['Tatal Staffed Beds', 'Covid Patients', 'Available Beds'])
fig.savefig('MetroHospitalBedsPatients.png')
```
