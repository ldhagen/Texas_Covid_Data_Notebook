```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
metro_pd = pd.read_csv('../CoVID-19_Dashboard_Data.csv')
```

```python
b = a.loc[a.index.dropna()]
```

```python
type(metro_pd.Date[0])
```

```python
c = b['DBCTestPositive'].dropna()
```

```python
type(c.index[0])
```

```python
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%b %Y')
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
plt.xlabel('Date',color='black')
plt.ylabel('Positive Tests',color='black')
plt.title('Daily Metro Bexar Positive Tests')
plt.bar(c.index,c)
#plt.plot(c)
fig.savefig('MetroDailyPositiveTests.png')
```

```python
d = b['DBCLabTests'].dropna()
```

```python
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%b %Y')
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
plt.xlabel('Date',color='black')
plt.ylabel('Daily Tests',color='black')
plt.title('Metro Bexar Daily Tests')
plt.bar(d.index,d)
#plt.plot(d)
fig.savefig('MetroDailyTests.png')
```

```python
e = pd.concat([c, d], axis=1)
```

```python
f = e.loc[e.index.dropna()]
```
