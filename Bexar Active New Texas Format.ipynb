```python
dated_dict = {}
count = 0
for x in a.T.Bexar[1:].index:
    month_temp = int(x[13:15])
    day_temp = int(x[16:18])
    if month_temp == 12:
        year_temp = 2020
    else:
        year_temp = 2021
    print(x)
    print(year_temp, month_temp, day_temp)
    dated_dict[datetime.datetime(year_temp, month_temp, day_temp)] = a.T.Bexar[count + 1]
    count += 1   
    
```

```python
dated_dict
```

```python
dated_df = pd.DataFrame(dated_dict.items(), columns=['Date', 'Total Cases'])
```

```python
dated_df
```

```python
c = dated_df.set_index('Date')
```

```python
c_7day_moving= c.rolling(7).mean().shift()
```

```python
import datetime
import matplotlib.dates as mdates
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%b %Y')
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
plt.xlabel('Date',color='black')
plt.ylabel('Active Cases',color='black')
plt.plot(c_7day_moving,'ro-')
plt.plot(c)
plt.legend(['7 Day Rolling Average','Daily Active Cases'])
plt.title('Bexar County Active COVID19 Estimated Cases From DSHS Texas Their New Format')
plt.savefig('BexarTotal.png')
```
