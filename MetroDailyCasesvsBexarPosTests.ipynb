```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
metro_pd = pd.read_csv('../CoVID-19_Dashboard_Data.csv')
```

```python
type(metro_pd.Date[0])
```

```python
metro_pd.Date = pd.to_datetime(metro_pd.Date)
```

```python
a = metro_pd.set_index('Date')
```

```python
a.index
```

```python
b = a.loc[a.index.dropna()]
```

```python
l = b['ReportedOn'].dropna()
```

```python
m = b['DBCTestPositive'].dropna()
```

```python
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%b %Y')
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
plt.xlabel('Date',color='black')
plt.ylabel('Daily Cases/Daily Positive Tests',color='black')
plt.title('San Antonio Dialy Cases vs Daily Bexar Positive Tests')
#plt.bar(c.index,c)
plt.plot(l, 'ro-')
plt.bar(m.index,m)
ax.legend(['San Antonio Daily Cases', 'Daily Bexar Positive Tests'])
fig.savefig('SACasesvsBexarPosTests.png')
```
